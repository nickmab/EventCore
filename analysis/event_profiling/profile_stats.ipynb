```python
import json
import pandas as pd
import numpy as np
```

```python
with open('EventProfiler_original_perf.json', 'r') as f:
    x = json.load(f)
    df = pd.read_json(json.dumps(x['traceEvents']))
    # This has the effect of sorting by which functions were called in which order. 
    # The raw data is logged in order of whatever _returned_ last.
    # We use mergesort because it's a stable sort. 
    # This has the effect of sorting by i) ts ascending (time of original call)
    # and then ii) index descending (so guarantees you see the 'outer' function call
    # appear first in the dataframe, even if they were all called at the 'same' time).
    df = df.sort_index(ascending=False)
    df = df.reset_index()
    df = df.sort_values('ts', kind='mergesort')
```
